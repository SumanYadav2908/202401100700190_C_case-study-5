```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Read the given .xlsx file
file_path = 'Case_Study_5_B.xlsx'  # replace with your actual file path
df = pd.read_excel(file_path)

# Step 2: Use NumPy to convert marks to percentage (Max mark = 15)
marks = df['Marks'].to_numpy()
percentages = (marks / 15) * 100
df['Percentage'] = percentages

# Step 3: Categorize students
greater_75 = df[df['Percentage'] > 75]
between_60_75 = df[(df['Percentage'] >= 60) & (df['Percentage'] <= 75)]
less_60 = df[df['Percentage'] < 60]

# Step 4: Print categorized data
print("ðŸ”µ Students with > 75%:\n", greater_75)
print("\nðŸŸ¡ Students with 60%-75%:\n", between_60_75)
print("\nðŸ”´ Students with < 60%:\n", less_60)

# Step 5: Plot Histogram
# Define categories and their counts
categories = ['>75%', '60-75%', '<60%']
counts = [len(greater_75), len(between_60_75), len(less_60)]

# Assign numerical positions to categories for correct alignment
positions = [0, 1, 2]  # x-tick positions

plt.figure(figsize=(8, 5))
plt.bar(positions, counts, width=0.5, color='blue', edgecolor='black')

# Set correct labels and ticks
plt.xticks(positions, categories)
plt.xlabel('Percentage Category')
plt.ylabel('Number of Students')
plt.title('Histogram Plot')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Step 6: Plot Scatter Plot (x: index of student, y: percentage)
plt.figure(figsize=(8, 5))
plt.scatter(df.index, df['Percentage'], color='green')
plt.xlabel('Students')
plt.ylabel('Percentage')
plt.title('Scatter Plot')
plt.grid(True)
plt.show()
```
